```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = 10
omega = 2 * np.pi * f
mu0 = np.pi * 4e-7
sigma = 0.01

z = np.arange(-1000, 10001, 10)
```

```python
def h(omega, sigma, z):
    k = np.sqrt(-1j * omega * np.pi * 4e-7 * sigma)
    if z < 0:
        H = 1 + 1j * 0
    if z >= 0:
        H = np.exp(-1j * k * z)
    return H
```

```python
def e(omega, sigma, z):
    k = np.sqrt(-1j * omega * np.pi * 4e-7 * sigma)
    if z < 0:
        E = -1j * k / sigma * (1 - 1j * k * z)
    if z >= 0:
        E = -1j * k / sigma * np.exp(-1j * k * z)
    return E
```

```python
H = [h(omega, sigma, zz) for zz in z]
E = [e(omega, sigma, zz) for zz in z]
H0 = h(omega, sigma, 0.0)
E0 = e(omega, sigma, 0.0)
```

```python
plt.plot(z, np.real(E / E0), label='Re')
plt.plot(z, np.imag(E / E0), label='Im')
plt.xlabel('z in m')
plt.ylabel(r'$E(z)/E(0)$')
plt.title('Normalized electric field')
plt.grid(True)
plt.legend()
```

```python
plt.plot(z, np.real(H / H0), label='Re')
plt.plot(z, np.imag(H / H0), label='Im')
plt.xlabel('z in m')
plt.ylabel(r'$H(z)/H(0)$')
plt.title('Normalized magnetic field')
plt.grid(True)
plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(4, 6), layout='constrained')
ax.plot(np.real(H / H0), z)
ax.invert_yaxis()
```
