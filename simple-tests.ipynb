```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sympy import *
```

```python
t, sigma, epsilon, rho0 = symbols('t sigma epsilon rho_0', real=True)
rhoE = symbols('rho_E', real=True)

ex = rhoE - rho0 * exp(-sigma / epsilon * t)
tt = solve(ex, t)
```

```python
tt = tt[0]
```

```python
print(latex(tt))
```

```python
sigma = 0.01
eps0 = 8.854e-12

def rho(t):
  return np.exp(-sigma/eps0 * t)

times = np.logspace(-13, -7, num=71, endpoint=True)

plt.semilogx(times, rho(times))
```
